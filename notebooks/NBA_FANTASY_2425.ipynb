### Getting stats from a certain date of the players rostered

```python
from nba_api.stats.endpoints import ScoreboardV2
from nba_api.stats.endpoints import BoxScoreTraditionalV2
```

```python
# 1. Fetch game IDs for a given date from ScoreboardV2
def fetch_game_ids_for_date(date):
    scoreboard = ScoreboardV2(game_date=date)
    games_data = scoreboard.get_dict()
    game_ids = [game[2] for game in games_data['resultSets'][0]['rowSet']]
    return game_ids
```

```python
# 2. Fetch individual player stats for each game from BoxScoreTraditionalV2
def fetch_player_stats_for_games(game_ids, league_players):
    all_player_stats = []
    
    for game_id in game_ids:
        # Fetch box score data for each game
        box_score = BoxScoreTraditionalV2(game_id=game_id)
        stats_data = box_score.player_stats.get_dict()  # Retrieve player stats from the response
        
        # Extract relevant player stats data, including 3PM (Three-Point Made) and MIN (Minutes)
        player_stats = pd.DataFrame(stats_data['data'], columns=stats_data['headers'])
        
        # Check for column presence
        required_columns = ['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ID', 'MIN', 'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA', 'REB', 'AST', 'STL', 'BLK', 'PTS']
        missing_columns = [col for col in required_columns if col not in player_stats.columns]
        if missing_columns:
            print(f"Warning: Missing columns {missing_columns} for game {game_id}. Skipping.")
            continue
        
        # Filter only players within the league_players DataFrame
        player_stats = player_stats[player_stats['PLAYER_ID'].isin(league_players['nba_id'])]
        
        # Append the player stats for this game to the list
        all_player_stats.append(player_stats)
    
    # Concatenate all player stats into a single DataFrame
    if all_player_stats:
        full_stats_df = pd.concat(all_player_stats, ignore_index=True)
        return full_stats_df
    else:
        return pd.DataFrame()
```

```python
# 3. Prepare daily stats for each player in the league
def prepare_daily_stats(date, league_players):
    # Fetch game IDs for the specified date
    game_ids = fetch_game_ids_for_date(date)
    if game_ids:
        # Fetch player stats for these games
        player_stats_df = fetch_player_stats_for_games(game_ids, league_players)
        
        if not player_stats_df.empty:
            # Add `GAME_DATE` column
            player_stats_df['GAME_DATE'] = date
            
            # Convert `MIN` if in 'hours.minutes:seconds' format
            if player_stats_df['MIN'].dtype == 'object':
                player_stats_df['MIN'] = player_stats_df['MIN'].apply(
                    lambda x: (
                        float(x.split(':')[0]) + float(x.split(':')[1]) / 60
                        if ':' in x else float(x)
                    ) if pd.notna(x) else None
                )
            
            # Round `MIN` before calculating `PPM`
            player_stats_df['MIN'] = player_stats_df['MIN'].round()
            
            # Convert necessary columns to numeric
            numeric_cols = ['FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA', 'PTS', 'MIN']
            for col in numeric_cols:
                if col in player_stats_df:
                    player_stats_df[col] = pd.to_numeric(player_stats_df[col], errors='coerce')
            
            # Calculate percentages and PPM
            player_stats_df['FG%'] = player_stats_df['FGM'] / player_stats_df['FGA']
            player_stats_df['FT%'] = player_stats_df['FTM'] / player_stats_df['FTA']
            player_stats_df['3P%'] = player_stats_df['FG3M'] / player_stats_df['FG3A']
            player_stats_df['PPM'] = player_stats_df['PTS'] / player_stats_df['MIN']
            
            # Merge with league_players to add player details
            player_stats_df = player_stats_df.merge(
                league_players[['nba_id', 'player_name', 'team_code']],
                left_on='PLAYER_ID',
                right_on='nba_id',
                how='left'
            ).drop(columns=['nba_id'])
            
            # Filter and reorder columns
            columns_order = [
                'GAME_DATE', 'team_code', 'player_name', 'PLAYER_ID', 
                'FGM', 'FG%', 'FTM', 'FT%', 'FG3M', '3P%', 
                'REB', 'AST', 'STL', 'BLK', 'PTS', 'PPM', 'MIN'
            ]
            available_columns = [col for col in columns_order if col in player_stats_df.columns]
            daily_stats_df = player_stats_df[available_columns]
            
            return daily_stats_df
        else:
            print("No player stats available.")
            return None
    else:
        print("No games found on this date.")
        return None
```

```python
# 4. Run the function to get daily stats for the specified date
date = '2024-11-19'
daily_stats_df = prepare_daily_stats(date, league_players)
print(daily_stats_df.info())
print(daily_stats_df.head())
```

### Showing best performers

```python


# Function to analyze top individual performers
def analyze_top_performers(daily_stats_df, top_n=5):
    # Top performers based on different categories
    categories = ['PTS', 'REB', 'AST', 'STL', 'BLK', 'PPM']
    top_performers = {}
    
    for category in categories:
        # Get the top N performers for each category
        top_performers[category] = daily_stats_df.nlargest(top_n, category)[['player_name', 'team_code', category]]
    
    return top_performers

# Function to analyze team performance, including player count
def analyze_team_performance(daily_stats_df):
    # Group by team_code and sum up the relevant stats for each team
    team_stats = daily_stats_df.groupby('team_code').agg({
        'PTS': 'sum',
        'REB': 'sum',
        'AST': 'sum',
        'STL': 'sum',
        'BLK': 'sum',
        'FGM': 'sum',
        'FTM': 'sum',
        'FG3M': 'sum',
        'PLAYER_ID': 'count'  # Count the number of players who played that day
    }).reset_index()
    
    # Rename PLAYER_ID column to Players_Active
    team_stats.rename(columns={'PLAYER_ID': 'Players_Active'}, inplace=True)
    
    # Calculate total stats as the sum of all relevant categories for each team
    team_stats['Total_Stats'] = (team_stats['PTS'] + team_stats['REB'] + team_stats['AST'] +
                                 team_stats['STL'] + team_stats['BLK'] + team_stats['FGM'] +
                                 team_stats['FTM'] + team_stats['FG3M'])
    
    # Sort the teams based on Total_Stats
    team_stats = team_stats.sort_values(by='Total_Stats', ascending=False)
    
    return team_stats

# Function to apply a smooth color gradient, format numbers, and set caption
def highlight_category(df, date, category=None):
    # Determine title based on category
    title = f"Top {category} Performers for {date}" if category else f"Team Performance Comparison for {date}"
    
    # Apply a smooth color gradient to all numeric columns
    styled_df = df.style.background_gradient(
        cmap='Greens', 
        axis=0
    ).format(precision=2, na_rep='').format(
        '{:.0f}', subset=df.select_dtypes(include=['float', 'int']).columns
    ).set_caption(title)
    
    return styled_df

# Function to perform full analysis and display best performers and team performance
def full_daily_analysis(daily_stats_df):
    # Extract the date from the daily_stats_df
    date = daily_stats_df['GAME_DATE'].iloc[0]  # Assuming all rows have the same date
    
    # Analyzing the top individual performers
    top_performers = analyze_top_performers(daily_stats_df, top_n=5)
    
    # Analyzing the team performance
    team_performance = analyze_team_performance(daily_stats_df)
    
    # Highlight individual performance for each category
    highlighted_performers = {}
    for category, df in top_performers.items():
        highlighted_performers[category] = highlight_category(df, date, category)
    
    # Highlight team performance based on all metrics
    highlighted_team_performance = highlight_category(team_performance, date)
    
    return highlighted_performers, highlighted_team_performance

# Example usage:
top_performers, team_performance = full_daily_analysis(daily_stats_df)

# Displaying the styled DataFrames
for category, styled_df in top_performers.items():
    display(styled_df)
    
display(team_performance)
```
