```python
# --- boxscore_inspect_v2.py ---
def peek_box_scores(league, matchup_period=None, scoring_period=None, limit=3):
    # Call with correct parameter names
    bs = league.box_scores(matchup_period=matchup_period, scoring_period=scoring_period)
    print(f"Matchups returned: {len(bs)}")
    if not bs:
        return

    g = bs[0]
    print("Matchup fields:")
    for a in ["matchup_period", "scoring_period", "winner"]:
        print(f"  {a}: {getattr(g, a, None)}")

    # Helper to inspect lineups
    def sample_lineup(lines, label):
        print(f"\n{label} lineup sample:")
        for pl in lines[:limit]:
            stats = getattr(pl, "stats", {}) or {}
            raw = next(iter(stats.values()), {})  # unwrap '0' layer
            print(f"  {pl.name} | {pl.position} | pro={pl.proTeam}")
            print("   stat keys:", list(raw.keys())[:12])

    sample_lineup(getattr(g, "home_lineup", []) or [], "HOME")
    sample_lineup(getattr(g, "away_lineup", []) or [], "AWAY")
```

```python
# --- boxscores_category_extractor.py ---
import pandas as pd
from datetime import datetime
from pathlib import Path

def extract_boxscores_category_df(league, matchup_period=None, scoring_period=None):
    """
    Extract per-player category stats from ESPN fantasy basketball box scores.
    Works with leagues using category scoring.
    """
    bs = league.box_scores(matchup_period=matchup_period, scoring_period=scoring_period)
    rows = []

    for game in bs:
        for side, team_obj, lineup in [
            ("HOME", getattr(game, "home_team", None), getattr(game, "home_lineup", []) or []),
            ("AWAY", getattr(game, "away_team", None), getattr(game, "away_lineup", []) or []),
        ]:
            for pl in lineup:
                stats_outer = getattr(pl, "stats", {}) or {}
                raw = next(iter(stats_outer.values()), {})  # unwrap '0'
                totals = raw.get("total", {}) or {}

                # Defensive mapping: some may be missing
                rec = {
                    "matchup_period": getattr(game, "matchup_period", None),
                    "scoring_period": getattr(game, "scoring_period", None),
                    "side": side,

                    "team_id": getattr(team_obj, "team_id", None),
                    "team_abbrev": getattr(team_obj, "team_abbrev", None),

                    "player_id": getattr(pl, "playerId", None),
                    "player_name": getattr(pl, "name", None),
                    "lineup_slot": getattr(pl, "position", None),
                    "pro_team": getattr(pl, "proTeam", None),
                    "injury_status": getattr(pl, "injuryStatus", None),
                }

                # Add all available totals
                for k, v in totals.items():
                    rec[k] = v
                rows.append(rec)

    df = pd.DataFrame(rows)
    if df.empty:
        print("⚠️ No player stats found — check if games have been played yet.")
        return df

    # Normalize common columns if present
    rename_map = {
        "fieldGoalsMade": "FGM",
        "fieldGoalsAttempted": "FGA",
        "freeThrowsMade": "FTM",
        "freeThrowsAttempted": "FTA",
        "threePointersMade": "3PM",
        "rebounds": "REB",
        "assists": "AST",
        "steals": "STL",
        "blocks": "BLK",
        "turnovers": "TO",
        "points": "PTS",
    }
    df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})

    # Derive percentages if possible
    if {"FGM", "FGA"}.issubset(df.columns):
        df["FG%"] = df.apply(lambda r: (r.FGM / r.FGA) if r.FGA else None, axis=1)
    if {"FTM", "FTA"}.issubset(df.columns):
        df["FT%"] = df.apply(lambda r: (r.FTM / r.FTA) if r.FTA else None, axis=1)

    print(f"✅ Extracted {len(df)} player lines across {df['team_abbrev'].nunique()} teams.")
    return df


def save_daily_category_stats(df, base_dir="data/processed/daily_stats", date_str=None):
    date_str = date_str or datetime.now().date().isoformat()
    month = date_str[:7]
    out_dir = Path(base_dir) / month
    out_dir.mkdir(parents=True, exist_ok=True)
    path = out_dir / f"basketball_stats_{date_str}.csv"
    df.to_csv(path, index=False)
    print(f"Saved: {path}")
    return str(path)
```

```python
def scoring_period_to_date(league, period):
    bs = league.box_scores(scoring_period=period)
    for g in bs:
        for pl in getattr(g, "home_lineup", []):
            stats_outer = getattr(pl, "stats", {}) or {}
            raw = next(iter(stats_outer.values()), {})
            date = raw.get("date")
            if date:
                return date  # 'YYYY-MM-DD'
    return None
```

```python
# === Cell 0: imports & paths ===
from pathlib import Path
from datetime import date, datetime, timedelta
from time import sleep

import json
import pandas as pd

from thefuzz import process
from nba_api.stats.static import players as nba_players_static
from nba_api.stats.endpoints import ScoreboardV2
from zoneinfo import ZoneInfo  # Python 3.9+
import unicodedata


# Local timezone (RD)
TZ_RD = ZoneInfo("America/Santo_Domingo")
TZ_ET = ZoneInfo("America/New_York")
TZ_UTC = ZoneInfo("UTC")

# Caches
NBA_ID_CACHE_PATH = Path("data/processed/mappings/nba_id_cache.json")
NBA_ID_CACHE_PATH.parent.mkdir(parents=True, exist_ok=True)

def normalize_name(name):
    """Normalize a name by removing accents."""
    return unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode('ASCII')

def build_nba_name_index():
    plist = nba_players_static.get_players()
    # Normalize names for matching
    by_name = {normalize_name(p["full_name"]): str(p["id"]) for p in plist}
    name_list = list(by_name.keys())
    return by_name, name_list


def _load_cache(path=NBA_ID_CACHE_PATH):
    return {k: str(v) for k, v in json.loads(path.read_text()).items()} if path.exists() else {}

def _save_cache(cache, path=NBA_ID_CACHE_PATH):
    path.write_text(json.dumps({k: str(v) for k, v in cache.items()}, ensure_ascii=False, indent=2))

def build_nba_name_index():
    plist = nba_players_static.get_players()
    # Normalize names for matching
    by_name = {normalize_name(p["full_name"]): str(p["id"]) for p in plist}
    name_list = list(by_name.keys())
    return by_name, name_list

def fuzzy_resolve(name, name_list, threshold=90):
    # Normalize the input name
    normalized_name = normalize_name(name)
    match, score = process.extractOne(normalized_name, name_list)
    return match if score >= threshold else None

def best_guess(name, name_list, k=3):
    normalized_name = normalize_name(name)
    return process.extract(normalized_name, name_list, limit=k)


```

```python
def _get(o, k, default=None): 
    return getattr(o, k, default)

def extract_league_players(league) -> pd.DataFrame:
    rows = []
    for t in league.teams:
        owners_data = _get(t, "owners", [])
        
        # Check if owners_data is a list of dicts
        if isinstance(owners_data, list) and owners_data and isinstance(owners_data[0], dict):
            # Extract the 'name' field from each owner dictionary
            owners = ", ".join(owner.get('name', 'Unknown') for owner in owners_data)
        else:
            # Assume owners_data is already a list of strings or a string
            owners = ", ".join(owners_data) if isinstance(owners_data, list) else owners_data

        for p in t.roster:
            rows.append({
                "team_id": _get(t, "team_id"),
                "team_abbrev": _get(t, "team_abbrev"),
                "team_name": _get(t, "team_name"),
                "player_id": _get(p, "playerId"),      # ESPN fantasy player id
                "player_name": _get(p, "name"),
                "pro_team": _get(p, "proTeam"),        # e.g., BOS, LAL
                "lineup_slot": _get(p, "position"),
            })
            
    df = pd.DataFrame(rows).drop_duplicates(subset=["player_id"]).reset_index(drop=True)
    return df

# Usage
league_players = extract_league_players(league)
league_players.head()
```

```python
# === Cell 2: map NBA_ID with cache + fuzzy ===
def map_nba_ids(league_players: pd.DataFrame) -> pd.DataFrame:
    cache = _load_cache()
    by_name, name_list = build_nba_name_index()
    out = league_players.copy()

    def resolve_id(row):
        key = f'{row["player_name"]}|{row.get("pro_team","")}'
        if key in cache:
            return cache[key]

        pid = by_name.get(row["player_name"])
        if pid:
            cache[key] = pid
            return pid

        match = fuzzy_resolve(row["player_name"], name_list, threshold=90)
        if match:
            pid = by_name[match]
            cache[key] = pid
            return pid

        return None

    out["nba_player_id"] = out.apply(resolve_id, axis=1).astype(str)
    _save_cache(cache)
    return out

league_players = map_nba_ids(league_players)
print("Mapped NBA IDs:", league_players["nba_player_id"].notna().sum(), "/", len(league_players))
league_players.head(20)
```

```python
from datetime import datetime, date
from zoneinfo import ZoneInfo
from pathlib import Path
import json
import pandas as pd
import requests

TZ_RD  = ZoneInfo("America/Santo_Domingo")
TZ_ET  = ZoneInfo("America/New_York")
TZ_UTC = ZoneInfo("UTC")

CACHE_DIR = Path("data/processed/misc")
CACHE_DIR.mkdir(parents=True, exist_ok=True)
SCHEDULE_CACHE = CACHE_DIR / "scheduleLeagueV2.json"  # temporada completa
```

```python
from nba_api.live.nba.endpoints import scoreboard as live_scoreboard

def first_last_tip_today_live():
    sb = live_scoreboard.ScoreBoard()  # solo hoy
    data = sb.get_dict().get("scoreboard", {})
    games = data.get("games", [])
    rows = []
    for g in games:
        # preferimos gameTimeUTC
        gt_utc = g.get("gameTimeUTC")
        if gt_utc and gt_utc.endswith("Z"):
            tip_utc = datetime.fromisoformat(gt_utc.replace("Z", "+00:00"))
        else:
            # fallback: '7:00 PM ET'
            game_et = g.get("gameEt")
            if not game_et:
                continue
            t = datetime.strptime(game_et.replace("ET","").strip(), "%I:%M %p")
            today = date.fromisoformat(data["gameDate"])
            tip_utc = datetime(today.year, today.month, today.day, t.hour, t.minute, tzinfo=TZ_ET).astimezone(TZ_UTC)

        rows.append({
            "GAME_ID": g.get("gameId"),
            "TIP_UTC": tip_utc,
            "TIP_ET":  tip_utc.astimezone(TZ_ET),
            "TIP_RD":  tip_utc.astimezone(TZ_RD),
        })
    if not rows:
        return None
    df = pd.DataFrame(rows).sort_values("TIP_UTC")
    first = df.iloc[0]; last = df.iloc[-1]
    return {
        "count_games": int(len(df)),
        "first": {k: (v.isoformat() if hasattr(v, "isoformat") else v) for k,v in first.items()},
        "last":  {k: (v.isoformat() if hasattr(v, "isoformat") else v) for k,v in last.items()},
        "source": "live_today"
    }

# Prueba:
first_last_tip_today_live()
```

```python
def load_season_schedule_from_cdn(season: int) -> dict:
    """
    Descarga (o lee del caché) el schedule de la temporada con UTCs reales por juego.
    Ej: season=2025 para 2025-26.
    """
    if SCHEDULE_CACHE.exists():
        try:
            return json.loads(SCHEDULE_CACHE.read_text())
        except Exception:
            pass  # si el caché está corrupto, re-descargamos

    # URL pública estable del calendario de la temporada
    # Nota: este endpoint publica la temporada completa (pre, reg y post) con 'gameDateTimeUTC'
    url = "https://cdn.nba.com/static/json/staticData/scheduleLeagueV2.json"
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    data = r.json()
    SCHEDULE_CACHE.write_text(json.dumps(data, ensure_ascii=False))
    return data

def first_last_tip_for_date_from_schedule(day: date, season_for_cache: int = None):
    """
    Busca en el schedule los partidos del día y retorna primer/último tip.
    No depende del ScoreboardV2 (que te ponía 00:00).
    """
    # season_for_cache no es obligatorio; usamos el JSON que trae la temporada en curso.
    sched = load_season_schedule_from_cdn(season_for_cache or day.year)
    gamesets = sched.get("leagueSchedule", {}).get("gameDates", [])
    rows = []
    wanted = day.isoformat()  # 'YYYY-MM-DD'

    for gd in gamesets:
        if gd.get("gameDate") != wanted:
            continue
        for g in gd.get("games", []):
            # gameDateTimeUTC viene como '2025-10-22T23:00:00Z'
            gt_utc = g.get("gameDateTimeUTC")
            if not gt_utc:
                continue
            tip_utc = datetime.fromisoformat(gt_utc.replace("Z","+00:00"))
            rows.append({
                "GAME_ID": g.get("gameId"),
                "TIP_UTC": tip_utc,
                "TIP_ET":  tip_utc.astimezone(TZ_ET),
                "TIP_RD":  tip_utc.astimezone(TZ_RD),
            })

    if not rows:
        return None

    df = pd.DataFrame(rows).sort_values("TIP_UTC")
    first = df.iloc[0]; last = df.iloc[-1]
    return {
        "count_games": int(len(df)),
        "first": {k: (v.isoformat() if hasattr(v, "isoformat") else v) for k,v in first.items()},
        "last":  {k: (v.isoformat() if hasattr(v, "isoformat") else v) for k,v in last.items()},
        "source": "cdn_schedule"
    }

# Pruebas:
 # temporada pasada
```

```python
def get_first_last_tip(day: date):
    today = date.today()
    if day == today:
        # Primero intentamos el live de hoy
        live = first_last_tip_today_live()
        if live: 
            return live
        # Si justo falla, usamos el schedule (también sirve para hoy)
        sch = first_last_tip_for_date_from_schedule(day)
        if sch:
            return sch
        # Último fallback: 19:00 RD
        dt_rd = datetime(day.year, day.month, day.day, 19, 0, tzinfo=TZ_RD)
        return {
            "count_games": 0,
            "first": {"GAME_ID":"", "TIP_RD": dt_rd.isoformat(), "TIP_ET": dt_rd.astimezone(TZ_ET).isoformat(), "TIP_UTC": dt_rd.astimezone(TZ_UTC).isoformat()},
            "last":  {"GAME_ID":"", "TIP_RD": dt_rd.isoformat(), "TIP_ET": dt_rd.astimezone(TZ_ET).isoformat(), "TIP_UTC": dt_rd.astimezone(TZ_UTC).isoformat()},
            "source": "default_7pm_RD"
        }
    else:
        # Fecha histórica o futura → schedule
        sch = first_last_tip_for_date_from_schedule(day)
        if sch:
            return sch
        # Fallback seguro
        dt_rd = datetime(day.year, day.month, day.day, 19, 0, tzinfo=TZ_RD)
        return {
            "count_games": 0,
            "first": {"GAME_ID":"", "TIP_RD": dt_rd.isoformat(), "TIP_ET": dt_rd.astimezone(TZ_ET).isoformat(), "TIP_UTC": dt_rd.astimezone(TZ_UTC).isoformat()},
            "last":  {"GAME_ID":"", "TIP_RD": dt_rd.isoformat(), "TIP_ET": dt_rd.astimezone(TZ_ET).isoformat(), "TIP_UTC": dt_rd.astimezone(TZ_UTC).isoformat()},
            "source": "default_7pm_RD"
        }

# Ejemplos:
get_first_last_tip(date.today())
get_first_last_tip(date(2024,10,22))
```
